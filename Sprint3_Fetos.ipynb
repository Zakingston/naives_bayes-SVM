## Train y test


```python
# Vamos a crear los conjuntos de entrenamiento y test con 60-40
x = data.drop("Target", axis=1)
y = data["Target"]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.40, random_state = 0)

```

Intento de escalar las variables con normalizaciÃ³n y estandarizaciÃ³n de las variables.

## Modelo

```python
# Vamos a modelar con Gaussian de Naive Bayes
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
modelNB = gnb.fit(x_train, y_train)
y_pred_train = modelNB.predict_proba(x_train)
y_pred_test = modelNB.predict_proba(x_test)
# predict_proba nos devuelve la predicciÃ³n en probabilidad
```

```python
print (y_pred_train)
```

```python
print (y_pred_test)
```

## ValidaciÃ³n del modelo con curva ROC



```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve, auc

fpr_train, tpr_train, _ = roc_curve(y_train, y_pred_train[:, 1])
roc_auc_train = auc(fpr_train, tpr_train)

fpr_test, tpr_test, _ = roc_curve(y_test, y_pred_test[:, 1])
roc_auc_test = auc(fpr_test, tpr_test)

# GrÃ¡fica
plt.figure()
lw = 2

plt.plot(fpr_train, tpr_train, color="darkblue", lw=lw, label = "Curva ROC entrenamiento(area = %0.2f)"% roc_auc_train,)
plt.plot(fpr_test, tpr_test, color="orange", lw=lw, label="Curva ROC test (area = %0.2f)"% roc_auc_test,)

plt.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC")
plt.legend(loc="lower right")
plt.show()
```

Como podemos ver en la grÃ¡fica, obtenemos un Ã¡rea bajo la curva ROC del 0,95 en entrenamiento y del 0,93 en test. El sobreajuste es mÃ­nimo. He intentando escalar las variables estandarizando y normalizando las variables sin Ã©xito. AÃºn asÃ­ son unos resultados muy buenos.

## Support Vector Machine

### Modelo dejando los valores por defecto

```python
from sklearn import svm
from sklearn.svm import SVC
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.50, random_state = 0)
```

```python
# SVM Lineal dejando los valores por defecto excepto la probabilidad para que nos deje validar el modelo con la curva ROC
svmLineal = svm.SVC(probability=True)
```

```python
# Modelo
svmLineal.fit(x_train, y_train)
```

```python
# Validacion del modelo
y_proba_train_svm = svmLineal.predict_proba(x_train)
y_proba_test_svm = svmLineal.predict_proba(x_test)
y_proba_test_svm
```

```python
# Train
fpr_train_svm, tpr_train_svm, _ = roc_curve(y_train, y_proba_train_svm[:, 1])
roc_auc_train_svm = auc(fpr_train_svm, tpr_train_svm)
# Test
fpr_test_svm, tpr_test_svm, _ = roc_curve(y_test, y_proba_test_svm[:, 1])
roc_auc_test_svm = auc(fpr_test_svm, tpr_test_svm)
```

```python
# GrÃ¡fica
plt.figure()
lw = 2
plt.plot(fpr_train_svm, tpr_train_svm, color = "blue", lw=lw, label = "Curva ROC entrenamiento (area = %0.2f)" % roc_auc_train_svm,)
plt.plot(fpr_test_svm, tpr_test_svm, color = "orange", lw=lw, label = "Curva ROC test (area =%0.2f)" % roc_auc_test_svm,)

plt.plot([0, 1], [0, 1], color = "navy", lw=lw, linestyle = "--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC")
plt.legend(loc="lower right")
plt.show()
```

Observamos un poco de sobreentrenamiento del modelo, lo comparamos a la curva que obtuvimos con Naive Bayes y luego lo intentamos ajustar.

### CompraciÃ³n con el modelo de Naive Bayes que obtuvimos

```python
# ComparaciÃ³n de curvas Navie Bayes y Support Vector Machine
plt.figure()
lw = 2
plt.plot(fpr_test, tpr_test, color="orange", lw=lw, label="Curva ROC Naive Bayes (area = %0.2f)"% roc_auc_test,)
plt.plot(fpr_test_svm, tpr_test_svm, color = "yellow", lw=lw, label = "Curva ROC SVM (area =%0.2f)" % roc_auc_test_svm,)

plt.plot([0, 1], [0, 1], color = "navy", lw=lw, linestyle = "--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC")
plt.legend(loc="lower right")
plt.show()
```

Comparando la curva ROC y el AUC para test de Naive Bayes y SVM, observamos que NB se ajusta mÃ¡s a 1, es decir, clasifica mejor que SVM. Esto pasarÃ­a dejando los valores por defecto, pero y si... Intentamos ajustar mÃ¡s nuestros hiperparÃ¡metros en SVM? Vamos a ello!!

### FunciÃ³n GridSearchCV

```python
# Vamos a aplicar la funciÃ³n GridSearchCV para buscar los mejores hiperparÃ¡metros
from sklearn.model_selection import GridSearchCV

# HiperparÃ¡metros
param_grid= [
            {"kernel":["rbf"], "gamma": [1e-3, 1e-4], "C": [0.1,1,10,20]},
            {"kernel":["linear"], "C": [0.1,1,10,20]},
            {"kernel":["poly"], "C": [0.1,1,10,20], "degree" : [2,3] }
]

grid = GridSearchCV(
      estimator = SVC(),
      param_grid = param_grid,
      scoring = "roc_auc",
      n_jobs = -1,
      cv = 3,
      verbose = 0,
      return_train_score = True
)
```

```python
grid.fit(X = x_train, y = y_train)
```

```python
# Obtenemos los mejores parÃ¡metros
print(grid.best_params_)
print(grid.best_estimator_)
```

### Volvemos a crear el modelo SVM con los mejores parÃ¡metros

Volvemos a crear nuestro modelo SVM con los hiperparÃ¡metros obtenidos en la funciÃ³n GridSearchCV.

```python
# HiperparÃ¡metros
svmLineal = svm.SVC(C = 0.1,
                    kernel = "linear",
                    probability=True
                    )
```

```python
# Modelo
svmLineal.fit(x_train, y_train)
```

```python
# ValidaciÃ³n del modelo
y_proba_train_svm = svmLineal.predict_proba(x_train)
y_proba_test_svm = svmLineal.predict_proba(x_test)
y_proba_test_svm
```

```python
# Train
fpr_train_svm, tpr_train_svm, _ = roc_curve(y_train, y_proba_train_svm[:, 1])
roc_auc_train_svm = auc(fpr_train_svm, tpr_train_svm)
# Test
fpr_test_svm, tpr_test_svm, _ = roc_curve(y_test, y_proba_test_svm[:, 1])
roc_auc_test_svm = auc(fpr_test_svm, tpr_test_svm)
```

```python
# GrÃ¡fica
plt.figure()
lw = 2
plt.plot(fpr_train_svm, tpr_train_svm, color = "blue", lw=lw, label = "Curva ROC entrenamiento (area = %0.2f)" % roc_auc_train_svm,)
plt.plot(fpr_test_svm, tpr_test_svm, color = "orange", lw=lw, label = "Curva ROC test (area =%0.2f)" % roc_auc_test_svm,)

plt.plot([0, 1], [0, 1], color = "navy", lw=lw, linestyle = "--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC")
plt.legend(loc="lower right")
plt.show()
```
